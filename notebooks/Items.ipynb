```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_path = '../input/data_v13.xlsx'
df = pd.read_excel(data_path)
```

```python
# Item 5
data8 = df.copy()

def calculate_item_5(row):
    if pd.isna(row['Gender']) or pd.isna(row['BMI']) or pd.isna(row['FORCEPREHENSION']):
        return np.nan

    if row['Gender'] == 1:
        if row['BMI'] <= 24:
            cutoff = 29
        elif 24.1 <= row['BMI'] <= 26:
            cutoff = 30
        elif 24.1 <= row['BMI'] <= 28:
            cutoff = 30
        else:                       # row['BMI'] > 28
            cutoff = 32
    else:                           # Gender == 0
        if row['BMI'] <= 23:
            cutoff = 17
        elif 23.1 <= row['BMI'] <= 26:
            cutoff = 17.3
        elif 26.1 <= row['BMI'] <= 29:
            cutoff = 18
        else:                       # row['BMI'] > 29
            cutoff = 21

    return 1 if row['FORCEPREHENSION'] > cutoff else 0

data8['Item_5_grip'] = data8.apply(calculate_item_5, axis=1)
data8.head()
```

```python
# 'Fried_Score' is the sum of the 5 items (Item_1, Item_2, Item_3, Item_4, Item_5). If one of the items is NaN, then the 'Fried_Score' is the sum of the other items.
data8 = df.copy()
data8['Fried_Score'] = data8[['Item_1', 'Item_2', 'Item_3', 'Item_4', 'Item_5']].dropna().sum(axis=1, min_count=1)
data8.head()
```

```python
# 'Fried_State' is 1 if 'Fried_Score' >= 3, else 0
data8['Fried_State_1'] = data8['Fried_Score'].apply(lambda x: 1 if x >= 3 else 0)
data8.head()
```
