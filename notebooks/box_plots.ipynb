```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_path = '../input/data_v13.xlsx'
data = pd.read_excel(data_path)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty_Score', y='Fried_Score', data=data[data.Frailty_Score != 5])
plt.title('Box Plot of Fried Score vs Frailty Score')
plt.xlabel('Frailty Score')
plt.ylabel('Fried Score')
plt.show()
```

```python
# Box plots for comparing the distribution of the target variables except for far values of x=5
sns.set(style="whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.boxplot(x='Frailty_Score', y='Fried_Score', data=data[data.x != 5], ax=ax[0])
sns.boxplot(x='y', y='x', data=data[data.x != 5], ax=ax[1])
ax[1].set(xscale="log")
plt.tight_layout()
plt.show()
```

```python
# Box plots for comparing the distribution of the target variables
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fried_Score', y='Frailty_Score', data=data[data.Fried_Score != 0])
plt.title('Box Plot of Frailty Score vs Fried Score')
plt.xlabel('Fried Score')
plt.ylabel('Frailty Score')
plt.show()
```
